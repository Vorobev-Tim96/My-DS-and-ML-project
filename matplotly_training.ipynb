```python
import pandas as pd
import numpy as np
import seaborn as sns
```

```python
import matplotlib.pyplot as plt
```

```python
from sklearn.datasets import make_moons
```

```python
features, target = make_moons(n_samples=100, noise=0.1, random_state=42)
```

```python
target
```

```python
fig, ax= plt.subplots(figsize=(16, 7), nrows=1, ncols=2, gridspec_kw={'width_ratios': [1, 1]})
sns.scatterplot(x=features[:,0], y=features[:,1], ax=ax[0])
sns.scatterplot(x=features[:,0], y=features[:,1], hue=target, ax=ax[1])
```

```python
df = pd.read_csv('dataset.csv')
df.head()
```

```python
df['University Rating'].value_counts()
```

```python
fig, axes = plt.subplots(figsize=(15, 6))
counts = df['University Rating'].value_counts()
sns.countplot(data=df, y='University Rating', ax=axes, order=counts.index, palette='magma')
```

```python
fig, axes = plt.subplots(figsize=(20, 5), nrows=1, ncols=5)
for i in range(1, 6):
  uni_rating = df[df['University Rating']==i]
  sns.histplot(uni_rating['GRE Score'], ax=axes[i-1], palette='plasma')
  axes[i - 1].set_title(f'GRE univercity rating {i}')
```

```python
fig, axes  = plt.subplots(1, 1, figsize=(10,10))
sns.kdeplot(data=df, x=df['TOEFL Score'], hue=df['University Rating'])
```

```python
sns.heatmap(df[['GRE Score', 'TOEFL Score',	'University Rating', 'Chance of Admit ']].corr())
```

```python
 fig, axes = plt.subplots(5, 1, figsize=(12, 9))
 for i in range(1, 6):
     df_type_rating = df[df['University Rating']==i]
     axes[i-1].set_ylabel('')
     axes[i-1].text(0.7, 0.5, f'University Rating: {i}', fontweight='bold', fontfamily='serif',
     fontsize=10, ha='right')
     sns.kdeplot(data=df_type_rating, x=df_type_rating['Chance of Admit '], fill=True, ax=axes[i-1])
```
