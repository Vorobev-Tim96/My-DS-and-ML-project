```python
import pandas as pd
import numpy as np
import warnings

import matplotlib.pyplot as plt
import numpy as np
from matplotlib import cm
from tqdm.auto import tqdm

warnings.filterwarnings("ignore")
%matplotlib inline
import torch.nn.functional as F
import torchvision
import torch as nn 

     
```

```python

import torch
from torch.nn.functional import conv2d
```

```python
transform = torchvision.transforms.Compose(
    [
     torchvision.transforms.ToTensor(),
     torchvision.transforms.Normalize((0.1307,), (0.3081,))
    ]
)
```

```python
mnist_train = torchvision.datasets.MNIST(
    "./mnist/",
    train=True,
    download=True,
    transform=transform
)
mnist_valid = torchvision.datasets.MNIST(
    "./mnist/",
    train=False,
    download=True,
    transform=transform 
)

train_dataloader = torch.utils.data.DataLoader(mnist_train, batch_size=64, shuffle=True)
valid_dataloader = torch.utils.data.DataLoader(mnist_valid, batch_size=64, shuffle=False)
```

```python
def train(model, optimazer, num_of_epoch):
  #train
  for epoch in range(num_of_epoch):
    for x_train, y_train in tqdm(train_dataloader):
      y_pred = model(x_train)
      loss = F.cross_entropy(y_pred, y_train)
      loss.backward()
      optimazer.step()
      optimazer.zero_grad()

    if epoch % 5 == 0:
      val_loss = []
      val_accuracy = []
      with torch.no_grad():
        for x_valid, y_valid in tqdm(valid_dataloader):
          y_pred = model(x_valid)
          val_loss.append(F.cross_entropy(y_pred, y_valid).numpy())
          val_accuracy.extend(((torch.argmax(y_pred, dim=-1) == y_valid).numpy().tolist()))
        
        print(f'Epoch: {epoch}, loss: {np.mean(val_loss)}, accuracy: {np.mean(val_accuracy)}')




```

```python

model = torch.nn.Sequential(
    torch.nn.Flatten(),             # превращаем картинку 28х28 в вектор размером 784
    torch.nn.Linear(28 * 28, 128),  # линейный слой, преобразующий вектор размера 784 в вектор размера 128
    torch.nn.ReLU(),                # нелинейность
    torch.nn.Linear(128, 10),       # линейный слой, преобразующий вектор размера 128 в вектор размера 10
)

optimizer = torch.optim.SGD(model.parameters(), lr=0.01) 
```

```python
train(model, optimizer, 50)
```

```python
model2 = torch.nn.Sequential(
    torch.nn.Conv2d(1, 10, 5),
    torch.nn.ReLU(),
    torch.nn.MaxPool2d(kernel_size=2),
    torch.nn.Conv2d(10, 20, 5),
    torch.nn.ReLU(),
    torch.nn.MaxPool2d(kernel_size=2),
    torch.nn.Flatten(),
    torch.nn.Linear(4*4*20, 128),
    torch.nn.ReLU(),
    torch.nn.Linear(128, 10),
)

optimizer = torch.optim.SGD(model2.parameters(), lr=0.01) 
```

```python
train(model2, optimizer, 50)
```
